##### 1 . BAR CHART

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel(r'C:\Users\arsha\Downloads\iris.xls')
data
```

```python


means = data.groupby('Classification')['SL'].mean()

plt.bar(means.index, means.values)
plt.xlabel('SL')
plt.ylabel('Mean Sepal Length')
plt.title('Mean Sepal Length by Classification')
plt.show()
```

```python
means = data.groupby('Classification')['SW'].mean()


plt.bar(means.index,means.values)
plt.xlabel('SW')
plt.ylabel('Mean Sepal width')
plt.title('Mean Sepal width by Classification')
plt.show()
```

```python
means = data.groupby('Classification')['PL'].mean()


plt.bar(means.index,means.values)
plt.xlabel('PL')
plt.ylabel('Mean petal length')
plt.title('Mean petal length by Classification')
plt.show()
```

```python
species_count = data['Classification'].value_counts()
plt.bar(species_count.index, species_count.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar plot of Species count')
plt.show()
```

### 2.SCATTER PLOT

```python
plt.scatter(data['SL'], data['SW'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
# Petal length vs petal width
```

```python
plt.scatter(data['PL'],data['PW'])
plt.xlabel('Petal_length')
plt.ylabel('Petal_width')
plt.title('Scatter plot of petal length vs petal width')
plt.show()
```

### 3. HISTOGRAM

```python
plt.hist(data['PL'], bins=20,color = 'blue')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

```python
#histogram of petal width
```

```python
plt.hist(data['PW'],bins = 20 ,color = 'red')
plt.xlabel('Petal width')
plt.ylabel('Frequency')
plt.title('Histogram of petal width')
plt.show()
```

```python
#histogram of sepal length
```

```python
plt.hist(data['SL'],bins = 20 ,color = 'green')
plt.xlabel('Sepal length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal length')
```

```python
#histogram of sepal width
plt.hist(data['SW'],bins = 20 ,color = 'brown')
plt.xlabel('sepal width')
plt.ylabel('Frequency')
plt.title('Histogram of sepal width')
plt.show()
```

### 4.BOX PLOT

```python
data.boxplot('PL', by='Classification', figsize=(8,6))
plt.xlabel('Classification')
plt.ylabel('Petal Length')
plt.title('Box plot of Petal Length by Classificaton')
plt.show()
```
